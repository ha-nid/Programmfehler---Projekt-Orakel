```python
#umständlicher Loop Sternzeichen Interactive
def Sternzeichen (Dummkopf,Steinbock,Wassermann,Fische,Widder,Stier,Zwillinge,Krebs,Löwe,Jungfrau,Waage,Skorpion,Schütze):
    ergebnis = ""
    if (Dummkopf):
        sleep(3)
        ergebnis = "Dummkopf"
    elif (Steinbock):
        sleep(3)
        ergebnis="Steinbock"
    elif (Wassermann):
        sleep(3)
        ergebnis= "Wassermann"
    elif (Fische):
        sleep(3)
        ergebnis="Fische"
    elif (Widder):
        sleep(3)
        ergebnis="Widder"
    elif (Stier):
        sleep(3)
        ergebnis="Stier"
    elif (Zwillinge):
        sleep(3)
        ergebnis="Zwillinge"
    elif (Krebs):
        sleep(3)
        ergebnis="Krebs"
    elif (Löwe):
        sleep(3)
        ergebnis="Löwe"
    elif (Jungfrau):
        sleep(3)
        ergebnis="Jungfrau"
    elif (Waage):
        sleep(3)
        ergebnis="Waage"
    elif (Skorpion):
        sleep(3)
        ergebnis="Skorpion"
    elif (Schütze):
        sleep(3)
        ergebnis="Schütze"
    else:
        sleep(4)
        ergebnis="Ich warte nicht gerne..."
    return (Dummkopf or Steinbock or Wassermann or Fische or Widder or Stier or Zwillinge or Krebs or Löwe or Jungfrau or Waage or Skorpion or Schütze)
interact(Sternzeichen, Dummkopf=Dropdown (value=None, options= ("Hallo","ne")),Steinbock=False,Wassermann=False,Fische=False,Widder=False,Stier=False,Zwillinge=False,Krebs=False,Löwe=False,Jungfrau=False,Waage=False,Skorpion=False,Schütze=False)
```

```python
#Loop Interactive Methode erste
def ErsterLoop (Nein,Ja):
    ergebnis = ""
    if (Nein):
        sleep(1)
        ergebnis = "Besser ist es, ich weiß sowieso alles am besten."
    elif (Ja):
        sleep(1)
        ergebnis = "Vielen Dank für nichts! Dann hättest du mich ja gar nicht erst belästigen müssen!"
    else:
        ergebnis="Ich warte nicht gerne!"
    return ergebnis
interact(ErsterLoop,Nein=Checkbox(value=None, description="Nein, ich möchte lieber dich fragen, oh allmächtiges Orakel!"), Ja=Checkbox(value=None, description="Ja, da hast du Recht, allwissendes Orakel!"))
sleep(2)
print('')
print("Da wir das jetzt geklärt haben, kanns ja weiter gehen.")
```
